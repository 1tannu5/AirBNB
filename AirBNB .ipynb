```python
import pandas as pd

# Load the CSV files
bangkok_data = pd.read_csv('C:\\Users\\stann\\OneDrive\\Documents\\datasets\\AIRBNB\\listings_bangkok.csv')
buenos_aires_data = pd.read_csv('C:\\Users\\stann\\OneDrive\\Documents\\datasets\\AIRBNB\\listings_buenos_aires.csv')
cape_town_data = pd.read_csv('C:\\Users\\stann\\OneDrive\\Documents\\datasets\\AIRBNB\\listings_cape_town.csv')
istanbul_data = pd.read_csv('C:\\Users\\stann\\OneDrive\\Documents\\datasets\\AIRBNB\\listings_istanbul.csv')

# Display the first few rows of each dataset
bangkok_data.head(), buenos_aires_data.head(), cape_town_data.head(), istanbul_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting style
sns.set(style="whitegrid")

# Helper function to plot room type distribution
def plot_room_type_distribution(data, city_name):
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x='room_type', palette='viridis')
    plt.title(f'Room Type Distribution in {city_name}')
    plt.xlabel('Room Type')
    plt.ylabel('Count')
    plt.show()
# Plot room type distribution for each city
plot_room_type_distribution(bangkok_data, 'Bangkok')
plot_room_type_distribution(buenos_aires_data, 'Buenos Aires')
plot_room_type_distribution(cape_town_data, 'Cape Town')
plot_room_type_distribution(istanbul_data, 'Istanbul')

# Helper function to plot average price of different room types
def plot_avg_price_per_room_type(data, city_name):
    plt.figure(figsize=(10, 6))
    avg_price = data.groupby('room_type')['price'].mean().reset_index()
    sns.barplot(data=avg_price, x='room_type', y='price', palette='viridis')
    plt.title(f'Average Price per Room Type in {city_name}')
    plt.xlabel('Room Type')
    plt.ylabel('Average Price')
    plt.show()


```

```python
# Plot average price of different room types for each city
plot_avg_price_per_room_type(bangkok_data, 'Bangkok')
plot_avg_price_per_room_type(buenos_aires_data, 'Buenos Aires')
plot_avg_price_per_room_type(cape_town_data, 'Cape Town')
plot_avg_price_per_room_type(istanbul_data, 'Istanbul')

# Helper function to plot distribution of availability
def plot_availability_distribution(data, city_name):
    plt.figure(figsize=(10, 6))
    sns.histplot(data['availability_365'], bins=30, kde=True, color='blue')
    plt.title(f'Distribution of Availability in {city_name}')
    plt.xlabel('Availability (days)')
    plt.ylabel('Count')
    plt.show()
```

```python

# Plot distribution of availability for each city
plot_availability_distribution(bangkok_data, 'Bangkok')
plot_availability_distribution(buenos_aires_data, 'Buenos Aires')
plot_availability_distribution(cape_town_data, 'Cape Town')
plot_availability_distribution(istanbul_data, 'Istanbul')

```

```python
# Helper function to plot average reviews per month
def plot_avg_reviews_per_month(data, city_name):
    plt.figure(figsize=(10, 6))
    avg_reviews = data.groupby('room_type')['reviews_per_month'].mean().reset_index()
    sns.barplot(data=avg_reviews, x='room_type', y='reviews_per_month', palette='viridis')
    plt.title(f'Average Reviews per Month in {city_name}')
    plt.xlabel('Room Type')
    plt.ylabel('Average Reviews per Month')
    plt.show()

# Plot average reviews per month for each city
plot_avg_reviews_per_month(bangkok_data, 'Bangkok')
plot_avg_reviews_per_month(buenos_aires_data, 'Buenos Aires')
plot_avg_reviews_per_month(cape_town_data, 'Cape Town')
plot_avg_reviews_per_month(istanbul_data, 'Istanbul')
```
